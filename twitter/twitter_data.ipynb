```python
import json

def search_word(word, lang='en'):
    global twitter
    # Get seach/tweets resource url
    url = 'https://api.twitter.com/1.1/search/tweets.json'
    params = {'q': word, 'lang': lang, 'count': '100'}
    req = twitter.get(url, params=params)
    if req.status_code == 200:
        data = json.loads(req.text)
        metadata = data['search_metadata']
        statuses = data['statuses']          
        return {"result":True, "metadata":metadata, "statuses":statuses}
    else: 
        print ("Error: %d" % req.status_code)
        return{"result":False, "status_code":req.status_code}
    
    return json.loads(req.text)
```

```python
import pymysql
import re

def create_db(db_info):
    connector = pymysql.connect(
        host = db_info["host"],
        user = db_info["user"],
        passwd = db_info["passwd"],
        charset = "utf8"
    )
    cursor = connector.cursor()
    sql = "CREATE DATABASE IF NOT EXISTS %s;" % db_info['db_name']
    cursor.execute(sql)
    connector.commit()
    cursor.close()
    connector.close()
    
def execute_sql(sql, db_info, is_commit = False):
    connector = pymysql.connect(
        host = db_info["host"],
        user = db_info["user"],
        passwd = db_info["passwd"],
        db = db_info['db_name'],
        charset = "utf8"
    )
    cursor = connector.cursor()
    cursor.execute(sql)
    if is_commit:
        connector.commit()
    cursor.close()
    connector.close()
    return True

def create_search_table(db_info):
    sql = "USE %s;" % db_info['db_name']
    execute_sql(sql, db_info, True)
    sql = """
    CREATE TABLE 
    IF NOT EXISTS
        Search(
            id BIGINT PRIMARY KEY AUTO_INCREMENT,
            datetime DATETIME,
            user_id BIGINT,
            followers_count BIGINT,
            friends_count BIGINT,
            tweet_id BIGINT,
            tweet TEXT,
            favorite_count BIGINT,
            retweet_count BIGINT,
            lang TEXT
        )
    ;
    """
    execute_sql(sql, db_info, is_commit=True)
    return True

def convert_search_dict(search_dict):
    search_date = datetime.strptime(search_dict['created_at'],'%a %b %d %H:%M:%S %z %Y')
    search_dict['created_at']= datetime.strftime(search_date, '%Y-%m-%d %H:%M:%S')
    search_dict['text'] = re.sub('[^a-zA-Z]', " ", search_dict['text'])
    return search_dict

def insert_into_search(db_info, search_dict):
    search_dict = convert_search_dict(search_dict)
    sql = """
    INSERT INTO
        Search
    VALUES(
        NULL, '%s', '%s', '%s', '%s', '%s', '%s', '%s', '%s', '%s'
        )
    ;
    """ %(
            search_dict['created_at'],
            search_dict['user']['id'],
            search_dict['user']['followers_count'],
            search_dict['user']['friends_count'],
            search_dict['id'],
            search_dict['text'],
            search_dict['favorite_count'],
            search_dict['retweet_count'],
            search_dict['lang']
        )
    execute_sql(sql, db_info, is_commit=True)
    return True
```
